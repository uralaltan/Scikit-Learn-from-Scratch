```python
def one_hot(df, columns, accuracy=100):
    new_df = df.copy()
    for i in columns:
        datas = df[i]
        new_df.drop(i, axis=1, inplace=True)
        classes = []
        temp = [(classes.append(i)) for i in datas[:accuracy] if i not in classes]
        num_class = len(classes)
        one_hot = np.zeros([datas.shape[0], num_class], dtype='int')
        for j, data in enumerate(datas):
            index = classes.index(data)
            one_hot[j][index] += 1
        temp_df = pd.DataFrame(data=one_hot, columns=classes)
        new_df = pd.concat([new_df, temp_df], axis=1)
        del temp
    return new_df
```

```python
def min_max_normalization(df, columns):
    df = pd.DataFrame({i: (df[i] - df[i].min()) / (df[i].max() + df[i].min()) if i in columns else df[i] for i in df.columns})
    return df

def log_normalization(df, columns):
    df = pd.DataFrame({i: (np.log(df[i])) if i in columns else df[i] for i in df.columns})
    return df

def z_score_normalization(df, columns):
    df = pd.DataFrame({i: (df[i] - df[i].mean()) / df[i].std() if i in columns else df[i] for i in df.columns})
    return df
```

### Logistic Regression ###

```python
def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))
```

```python
def cost(X, y, W, b, lambd=0): 
    m = X.shape[0]
    cost=0
    for i in range(m):
        prediction = sigmoid(np.dot(X[i], W) + b)
        cost += -y[i] * np.log(prediction) - (1 - y[i]) * np.log(1 - prediction)
    cost += lambd * np.sum(W**2) # L2 regularization
    return cost/(2*m)
```

```python
def gradient(X, y, W, b, lambd=0): 
    m, n = X.shape
    derivative_w = np.zeros(W.shape)
    derivative_b = 0.
    for i in range(m):
        predictions = sigmoid((np.dot(X[i], W) + b))
        loss = (predictions - y[i]) ** 2
        for j in range(n):
            derivative_w[j] += loss * X[i, j] + lambd * W[j] # L2 regularization
        derivative_b += predictions
    return derivative_w/m, derivative_b/m
```

```python
def gradient_descent(X, y, W, b, cost_function, gradient_function, alpha, num_iters, lambd=0): 
    m, n = X.shape
    for i in range(num_iters):
        derivative_w, derivative_b = gradient_function(X, y, W, b, lambd)  
        W = W - alpha * derivative_w               
        b = b - alpha * derivative_b               
        cost =  np.sum(cost_function(X, y, W, b, lambd))
        if i % (num_iters/10) == 0:
            print(f"Iteration {i:4}: Cost {float(cost):8.4f}   ")
    print(f"Iteration {num_iters:4}: Cost {float(cost):8.4f}   ")
    return W, b, lambd
```

```python
X = np.array(data.drop('cb_person_default_on_file', axis=1))
y = np.array(data['cb_person_default_on_file'])

X_train = X[2000:]
y_train = y[2000:]

X_test = X[:2000]
y_test = y[:2000]

weights = np.random.rand(X.shape[1]) -0.5
bias = np.random.rand(1) -0.5
```

```python
w, b, _ = gradient_descent(X_train, y_train, weights, bias, cost, gradient, 0.1, 20)
```

```python
w2, b2, _ = gradient_descent(X_train, y_train, w, b, cost, gradient, 0.04, 10)
```

```python
def accuracy(X, y, W, b, threshold=0.5):
    m, n = X.shape   
    prediction = np.zeros(m)
    for i in range(m):   
        model = np.dot(X[i], W) + b
        g = sigmoid(model)
        prediction[i] = 1 if g > threshold else 0
    prediction = np.array(prediction, dtype='int64')
    return (prediction == y).sum() / y.shape[0]
```

```python
accuracy(X_train, y_train, w2, b2, threshold=0.6)
```

```python
accuracy(X_test, y_test, w2, b2, threshold=0.6)
```
